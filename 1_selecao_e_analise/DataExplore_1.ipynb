# Outliers

```python
print(df['QT_SOLICITADA'])

print(df['QT_SOLICITADA'].max())

print(df['QT_SOLICITADA'].min())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 10))
plt.boxplot(df['QT_SOLICITADA'], vert=True)
plt.ylabel('QT_SOLICITADA')
plt.ylim(0, 10)  
plt.show()

# Calculando os quartis para QT_SOLICITADA
quartiles = df['QT_SOLICITADA'].quantile([0.25, 0.5, 0.75])

quartiles
```

```python
# Calculando o IQR
IQR = quartiles[0.75] - quartiles[0.25]

# Definindo os limites para outliers
lower_bound = quartiles[0.25] - 1.5 * IQR
upper_bound = quartiles[0.75] + 1.5 * IQR

# Identificando os outliers
outliers = df[(df['QT_SOLICITADA'] < lower_bound) | (df['QT_SOLICITADA'] > upper_bound)]

# Número de outliers
num_outliers = outliers.shape[0]

num_outliers, lower_bound, upper_bound
```

```python
# Identificando o DS_STATUS_ITEM para o valor mínimo de QT_SOLICITADA
status_min = df[df['QT_SOLICITADA'] == df['QT_SOLICITADA'].min()]['DS_STATUS_ITEM'].unique()

# Identificando o DS_STATUS_ITEM para o valor máximo de QT_SOLICITADA
status_max = df[df['QT_SOLICITADA'] == df['QT_SOLICITADA'].max()]['DS_STATUS_ITEM'].unique()

status_min, status_max
```

```python
# Filtrando os outliers de QT_SOLICITADA
outliers_data = df[df['QT_SOLICITADA'] > upper_bound]

# Verificando a distribuição do DS_STATUS_ITEM para os outliers
status_distribution = outliers_data['DS_STATUS_ITEM'].value_counts()

status_distribution
```

```python
# Filtrando os valores que não são outliers de QT_SOLICITADA
non_outliers_data = df[df['QT_SOLICITADA'] <= upper_bound]

# Verificando a distribuição do DS_STATUS_ITEM para os não outliers
non_status_distribution = non_outliers_data['DS_STATUS_ITEM'].value_counts()

non_status_distribution
```

# Formatação das datas

```python
from datetime import timedelta
d = timedelta(hours=2432857)
(d.days, d.seconds, d.microseconds)


anos = d.days / 365
print(anos)
```

```python
def jd_to_date(jd):
    # Algoritmo para converter Julian Date para data gregoriana
    # Referência: Jean Meeus, "Astronomical Algorithms", 1991
    Z = int(jd + 0.5)
    F = jd + 0.5 - Z
    if Z < 2299161:
        A = Z
    else:
        alpha = int((Z - 1867216.25) / 36524.25)
        A = Z + 1 + alpha - int(alpha / 4)
    
    B = A + 1524
    C = int((B - 122.1) / 365.25)
    D = int(365.25 * C)
    E = int((B - D) / 30.6001)
    
    # Calculando dia, mês e ano
    day = B - D - int(30.6001 * E) + F
    month = E - 1 if E < 13.5 else E - 13
    year = C - 4716 if month > 2.5 else C - 4715
    
    return year, month, int(day)

# Exemplo de uso
jd_value = 2440511  # Substitua pelo seu valor de JD, se necessário
date_result = jd_to_date(jd_value)
print(date_result)
```
